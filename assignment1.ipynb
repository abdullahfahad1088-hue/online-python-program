```python
# 7 write a program which takes 5 inputs from user for different subjects marks total it and generate mark sheets using grades
eng = int(input("Enter the English Marks: "))
urdu = int(input("Enter the Urdu Marks: "))
math = int(input("Enter the Math Marks: "))
pst = int(input("Enter the PST Marks: "))
comp = int(input("Enter the Computer Marks: "))
obtain = eng + urdu + math + pst + comp
print("Your Obtain Marks: ", obtain)
perc = obtain / 500 * 100
print("Your Percentage is : ", perc)
if perc >=80 and perc <=100:
    print("A-One Grade")
elif perc >= 70 and perc < 80:
    print("A Grade")
elif perc >= 60 and perc < 70:
    print("B Grade")
elif perc >= 50 and perc < 60:
    print("C Grade")
elif perc >= 40 and perc < 50:
    print("D Grade")
elif perc >= 33 and perc < 40:
    print("E Grade")
else:
    print("Fail")
```
