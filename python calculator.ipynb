```python
# Python calculator
num1 = float(input("Enter the first number: "))
operator = input("Enter an operator(+ - * /): ")
num2 = float(input("Enter the second number: "))


if operator == "+":
    result = num1 + num2
    print(result)
elif operator == "-":
    result = num1 - num2
    print(result)
elif operator == "*":
    result = num1 * num2
    print(result)
elif operator == "/":
    result = num1 / num2
    print(result)
else:
    print("Please insert correct operator")
```
